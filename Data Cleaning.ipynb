```python
#Load the dataset
file_path = 'encrypted_sales_data.xlsx'
data = pd.read_excel(file_path)
```

```python
import matplotlib.pyplot as plt

# Visualizing the distribution of 'Items.Purchase Price' and 'Total_GP'
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot for 'Items.Purchase Price'
axes[0].hist(data['Items.Purchase Price'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Items.Purchase Price')
axes[0].set_xlabel('Purchase Price')
axes[0].set_ylabel('Frequency')

# Plot for 'Total_GP'
axes[1].hist(data['Total_GP'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Distribution of Total_GP')
axes[1].set_xlabel('Total_GP')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### Items.Purchase Price: The distribution is right-skewed, suggesting that while most purchase prices are low, there are a few high values pulling the mean up. Using the median would be more appropriate for imputation to avoid the influence of these outliers.

#### Total_GP: This also appears right-skewed with some high-value outliers. Again, the median is likely more suitable for imputing missing values to represent a typical gross profit more accurately.

```python
#Cleaning the data
#Calculate median values for 'Items.Purchased Price and 'Total_GP'

purchase_price_median = data['Items.Purchase Price'].median()
total_gp_median = data['Total_GP'].median()

# Correcting the fillna usage to avoid FutureWarnings
data['Items.Purchase Price'] = data['Items.Purchase Price'].fillna(purchase_price_median)
data['Total_GP'] = data['Total_GP'].fillna(total_gp_median)

# Re-running the dropna operation for ID fields to refresh the dataset and confirm changes
data_cleaned = data.dropna(subset=['Invoice Items.Product ID', 'Items.Item ID'])

# Check the cleaned data again
data_cleaned.info()
```

```python
#Ckecking for missing values 
missing_values = data_cleaned.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Dropping the column 'Invoices.CRM Potential ID'
data_cleaned = data_cleaned.drop(columns=['Invoices.CRM Potential ID'])

# Check the dataframe structure again to confirm the column has been removed
data_cleaned.info()
```

```python
# Saving the cleaned dataset to a new Excel file
cleaned_file_path = 'cleaned_sales_data.xlsx'
data_cleaned.to_excel(cleaned_file_path, index=False)

cleaned_file_path
```
